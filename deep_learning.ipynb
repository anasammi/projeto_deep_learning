```python
from numpy import loadtxt
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
#carregando base de dados
dataset = loadtxt('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv', delimiter=',')

#separando os dados entre entradas(X) e saídas(y)
X= dataset[:,0:8]
y= dataset[:,8]

```

```python
#definindo modelo com Keras
#inicializando o modelo sequencial
model = Sequential()
#inicializando primeira camada com 12 neurônios, 8 entradas utilizando a função ReLU
model.add(Dense(12, input_dim=8, activation='relu'))
#inicializando segunda camada com 8 neurônios utilizando a função ReLU
model.add(Dense(8, activation='relu'))
#inicializando terceira camada (camada de saída) com 1 neurônio utilizando a função Sigmoid
model.add(Dense(1, activation='sigmoid'))
```

```python
#compilando Keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#treinando o modelo Keras
model.fit(X, y, epochs=150, batch_size=10)
```

```python
#avaliando o modelo Keras
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
